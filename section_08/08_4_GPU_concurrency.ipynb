## GPU Concurrency
* GPUs are designed to very efficiently do a lot of small tasks
* Initially for rendering individual pixels using a pipeline of operations
* Each pixel can be rendered somewhat independently, ie. they are trivially parallelizable
* Same approach is useful for doing large array operations

* We can use libraries like `Pytorch` to do large array operations using GPUs
* Data must be put into tensors which then get copied to GPU memory
* GPUs have special on-board memory that is faster but smaller than main memory and addressed separately
* Array operations then broken up into thousands of individual operations and executed

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
device = torch.device("cuda:0")

x, y = np.mgrid[0 : np.pi * 2 : np.pi / 100, 0 : np.pi * 2 : np.pi / 300]

a = torch.from_numpy(x).to(device)  # convert to tensors and
b = torch.from_numpy(y).to(device)  # copy to GPU memory

c = torch.cos(a) * torch.sin(b)  # execute operations in parallel
plt.imshow(c.cpu().data.numpy())  # copy back to main memory
```
